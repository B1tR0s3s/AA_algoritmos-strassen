```python
import numpy as np

# Matrix dimensions (A1: 10x30, A2: 30x5, A3: 5x60, A4: 60x20)
dims = [10, 30, 5, 60, 20]  # Matrix dimensions

# Number of matrices
n = len(dims) - 1

# Initialize cost matrix (m[i][j]) with large values
m = np.full((n, n), float('inf'))
s = np.zeros((n, n), dtype=int)  # To store split points

# Fill diagonal elements (single matrices require zero multiplication)
for i in range(n):
    m[i][i] = 0

# Apply Matrix Chain Multiplication DP algorithm
for chain_len in range(2, n + 1):  # Chain length (2 to n)
    for i in range(n - chain_len + 1):
        j = i + chain_len - 1
        for k in range(i, j):
            q = m[i][k] + m[k+1][j] + dims[i] * dims[k+1] * dims[j+1]
            if q < m[i][j]:
                m[i][j] = q
                s[i][j] = k

# Function to print optimal parenthesization
def print_parenthesization(s, i, j):
    if i == j:
        return f"A{i+1}"
    else:
        return f"({print_parenthesization(s, i, s[i][j])} × {print_parenthesization(s, s[i][j]+1, j)})"

# Get the optimal order
optimal_order = print_parenthesization(s, 0, n-1)
print("Optimal Parenthesization:", optimal_order)
print("Minimum Multiplications:", m[0][n-1])

# --- Hybrid Multiplication Execution ---
total_operations = 0

# Step 1: A3 × A4
step1_ops = dims[2] * dims[3] * dims[4]  # 5×60×20
print("Step 1: A3 × A4 =", step1_ops, "operations")
total_operations += step1_ops

# Step 2: A2 × (A3×A4)
step2_ops = dims[1] * dims[2] * dims[4]  # 30×5×20
print("Step 2: A2 × (A3×A4) =", step2_ops, "operations")
total_operations += step2_ops

# Step 3: A1 × (A2 × (A3 × A4))
step3_ops = dims[0] * dims[1] * dims[4]  # 10×30×20
print("Step 3: A1 × (A2 × (A3 × A4)) =", step3_ops, "operations")
total_operations += step3_ops

print("Total Multiplications Required:", total_operations)
```

```python
import numpy as np

# --- MATRIX CHAIN ORDER COMPUTATION ---
def matrix_chain_order(p):
    n = len(p) - 1  # Number of matrices
    m = np.full((n, n), float('inf'))  # Cost table
    s = np.zeros((n, n), dtype=int)  # Parenthesization table

    for i in range(n):
        m[i, i] = 0  # Cost of multiplying one matrix is zero

    for chain_length in range(2, n+1):  # Chain length from 2 to n
        for i in range(n - chain_length + 1):
            j = i + chain_length - 1
            for k in range(i, j):
                cost = m[i, k] + m[k+1, j] + p[i] * p[k+1] * p[j+1]
                if cost < m[i, j]:
                    m[i, j] = cost
                    s[i, j] = k

    return m, s

# --- STRASSEN'S MATRIX MULTIPLICATION ---
def strassen_matrix_multiply(A, B):
    """ Performs Strassen's matrix multiplication for square matrices of size 2^n × 2^n. """
    assert A.shape == B.shape, "Strassen's method requires square matrices of the same size."

    n = A.shape[0]
    if n == 1:  # Base case: 1x1 matrix
        return A * B

    mid = n // 2  # Split point

    # Divide matrices into quadrants
    A11, A12, A21, A22 = A[:mid, :mid], A[:mid, mid:], A[mid:, :mid], A[mid:, mid:]
    B11, B12, B21, B22 = B[:mid, :mid], B[:mid, mid:], B[mid:, :mid], B[mid:, mid:]

    # Compute the 7 Strassen's intermediate matrices
    M1 = strassen_matrix_multiply(A11 + A22, B11 + B22)
    M2 = strassen_matrix_multiply(A21 + A22, B11)
    M3 = strassen_matrix_multiply(A11, B12 - B22)
    M4 = strassen_matrix_multiply(A22, B21 - B11)
    M5 = strassen_matrix_multiply(A11 + A12, B22)
    M6 = strassen_matrix_multiply(A21 - A11, B11 + B12)
    M7 = strassen_matrix_multiply(A12 - A22, B21 + B22)

    # Compute result quadrants
    C11 = M1 + M4 - M5 + M7
    C12 = M3 + M5
    C21 = M2 + M4
    C22 = M1 - M2 + M3 + M6

    # Combine results
    C = np.vstack((np.hstack((C11, C12)), np.hstack((C21, C22))))
    return C

# --- HYBRID MATRIX CHAIN MULTIPLICATION ---
def hybrid_matrix_chain_multiplication(p, matrices):
    """ Computes optimal matrix chain multiplication using DP and applies Strassen's method where applicable. """
    m, s = matrix_chain_order(p)
    n = len(matrices)

    def multiply_recursive(i, j):
        if i == j:  # Base case: Single matrix
            return matrices[i]

        k = s[i, j]
        A = multiply_recursive(i, k)
        B = multiply_recursive(k + 1, j)

        # Use Strassen's method only for square matrices with size 2^n × 2^n
        if A.shape == B.shape and A.shape[0] % 2 == 0:
            return strassen_matrix_multiply(A, B)
        else:
            return np.dot(A, B)  # Standard multiplication

    return multiply_recursive(0, n - 1)

# --- EXAMPLE USAGE ---
if __name__ == "__main__":
    # Example Matrices
    A1 = np.random.randint(1, 10, (10, 30))
    A2 = np.random.randint(1, 10, (30, 5))
    A3 = np.random.randint(1, 10, (5, 60))
    A4 = np.random.randint(1, 10, (60, 20))

    matrices = [A1, A2, A3, A4]
    dimensions = [10, 30, 5, 60, 20]  # Dimensions of matrices

    # Compute optimal multiplication using hybrid approach
    result = hybrid_matrix_chain_multiplication(dimensions, matrices)
    print("Final Matrix Shape:", result.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import time
import psutil
import os
import gc

# Matrix sizes
matrix_sizes = np.array([64, 128, 256, 512, 1024])
STRASSEN_THRESHOLD = 128

process = psutil.Process(os.getpid())

def get_memory_mb():
    return process.memory_info().rss / (1024 * 1024)

def strassen(A, B):
    n = A.shape[0]
    if n <= STRASSEN_THRESHOLD:
        return A @ B
    mid = n // 2
    A11, A12, A21, A22 = A[:mid,:mid], A[:mid,mid:], A[mid:,:mid], A[mid:,mid:]
    B11, B12, B21, B22 = B[:mid,:mid], B[:mid,mid:], B[mid:,:mid], B[mid:,mid:]
    M1 = strassen(A11 + A22, B11 + B22)
    M2 = strassen(A21 + A22, B11)
    M3 = strassen(A11, B12 - B22)
    M4 = strassen(A22, B21 - B11)
    M5 = strassen(A11 + A12, B22)
    M6 = strassen(A21 - A11, B11 + B12)
    M7 = strassen(A12 - A22, B21 + B22)
    C11 = M1 + M4 - M5 + M7
    C12 = M3 + M5
    C21 = M2 + M4
    C22 = M1 - M2 + M3 + M6
    C = np.zeros((n, n), dtype=A.dtype)
    C[:mid,:mid], C[:mid,mid:], C[mid:,:mid], C[mid:,mid:] = C11, C12, C21, C22
    return C

standard_multiplication = []
strassen_algorithm = []
dp_mcm = []
hybrid_mcm_strassen = []
standard_memory = []
strassen_memory = []
dp_memory = []
hybrid_memory = []

for n in matrix_sizes:
    np.random.seed(42)
    A = np.random.rand(n, n).astype(np.float64)
    B = np.random.rand(n, n).astype(np.float64)

    # Standard multiplication
    gc.collect()
    mem_before = get_memory_mb()
    start = time.perf_counter()
    C_std = A @ B
    end = time.perf_counter()
    mem_after = get_memory_mb()
    standard_time = (end - start) * 1000
    standard_mem = max(mem_after - mem_before, 0)

    # Strassen algorithm
    gc.collect()
    mem_before = get_memory_mb()
    start = time.perf_counter()
    C_str = strassen(A, B)
    end = time.perf_counter()
    mem_after = get_memory_mb()
    strassen_time = (end - start) * 1000
    strassen_mem = max(mem_after - mem_before, 0)

    # DP MCM (standard multiply chain of 2 matrices = standard)
    gc.collect()
    mem_before = get_memory_mb()
    start = time.perf_counter()
    _ = A @ B
    end = time.perf_counter()
    mem_after = get_memory_mb()
    dp_time = (end - start) * 1000
    dp_mem = max(mem_after - mem_before, 0)

    # Hybrid MCM (uses Strassen for multiply step)
    gc.collect()
    mem_before = get_memory_mb()
    start = time.perf_counter()
    _ = strassen(A, B)
    end = time.perf_counter()
    mem_after = get_memory_mb()
    hybrid_time = (end - start) * 1000
    hybrid_mem = max(mem_after - mem_before, 0)

    standard_multiplication.append(standard_time)
    strassen_algorithm.append(strassen_time)
    dp_mcm.append(dp_time)
    hybrid_mcm_strassen.append(hybrid_time)
    standard_memory.append(standard_mem)
    strassen_memory.append(strassen_mem)
    dp_memory.append(dp_mem)
    hybrid_memory.append(hybrid_mem)

    del A, B, C_std, C_str
    gc.collect()

standard_multiplication = np.array(standard_multiplication)
strassen_algorithm = np.array(strassen_algorithm)
dp_mcm = np.array(dp_mcm)
hybrid_mcm_strassen = np.array(hybrid_mcm_strassen)
standard_memory = np.array(standard_memory)
strassen_memory = np.array(strassen_memory)
dp_memory = np.array(dp_memory)
hybrid_memory = np.array(hybrid_memory)

# Compute Efficiency Gain (%)
speedup = ((standard_multiplication - hybrid_mcm_strassen) / standard_multiplication) * 100
scalar_reduction = ((standard_multiplication - hybrid_mcm_strassen) / standard_multiplication) * 100
memory_reduction = ((standard_memory - hybrid_memory) / standard_memory) * 100

# Create a DataFrame for tabular representation
df = pd.DataFrame({
    "Matrix Size": matrix_sizes,
    "Exec Time Std (ms)": standard_multiplication,
    "Exec Time Strassen (ms)": strassen_algorithm,
    "Exec Time DP MCM (ms)": dp_mcm,
    "Exec Time Hybrid MCM (ms)": hybrid_mcm_strassen,
    "Memory Std (MB)": standard_memory,
    "Memory Strassen (MB)": strassen_memory,
    "Memory DP MCM (MB)": dp_memory,
    "Memory Hybrid MCM (MB)": hybrid_memory,
    "Speedup (%)": speedup,
    "Memory Reduction (%)": memory_reduction
})

print(df)

# --- Plot Execution Time ---
plt.figure(figsize=(8,5))
plt.plot(matrix_sizes, standard_multiplication, 'o-', label="Standard", color='orange')
plt.plot(matrix_sizes, strassen_algorithm, 's--', label="Strassen", color='red')
plt.plot(matrix_sizes, dp_mcm, '^-.', label="DP MCM", color='green')
plt.plot(matrix_sizes, hybrid_mcm_strassen, 'd:', label="Hybrid MCM", color='blue')
plt.xlabel("Matrix Size")
plt.ylabel("Execution Time (ms)")
plt.title("Execution Time Comparison")
plt.legend()
plt.grid()
plt.show()

# --- Plot Memory Usage ---
plt.figure(figsize=(8,5))
plt.plot(matrix_sizes, standard_memory, 'o-', label="Standard", color='orange')
plt.plot(matrix_sizes, strassen_memory, 's--', label="Strassen", color='red')
plt.plot(matrix_sizes, dp_memory, '^-.', label="DP MCM", color='green')
plt.plot(matrix_sizes, hybrid_memory, 'd:', label="Hybrid MCM", color='blue')
plt.xlabel("Matrix Size")
plt.ylabel("Memory Usage (MB)")
plt.title("Memory Usage Comparison")
plt.legend()
plt.grid()
plt.show()

# --- Plot Efficiency Gains ---
plt.figure(figsize=(8,5))
plt.plot(matrix_sizes, speedup, 'o-', label="Speedup (Std vs Hybrid)", color='blue')
plt.plot(matrix_sizes, memory_reduction, 's--', label="Memory Reduction", color='purple')
plt.xlabel("Matrix Size")
plt.ylabel("Efficiency Gain (%)")
plt.title("Efficiency Gains of Strassen over Standard")
plt.legend()
plt.grid()
plt.show()
```
